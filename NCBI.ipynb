```python
            
def gen(genasoc):
    lista=[]
    print("\n"+genasoc+"--------------------------------------------")
    genasoc=str(genasoc)
    genasoc=genasoc.replace(",","%2C").replace(" ","+").replace("\xa0","%C2%A0")
    genasoc="https://www.ncbi.nlm.nih.gov/gene/?term="+genasoc
    url=genasoc
    uclient=urlopen(url)
    pagehtml=uclient.read()
    uclient.close()
    pagesoup=soup(pagehtml, "html.parser")
    page=pagesoup.find("div",{"class":"ui-ncbigrid-inner-div"})
    locations=page.tbody.findAll("tr",{"class":"rprt"})
    for i in range(0,len(locations)):
        if locations[i].findAll("td")[2].text=="":
            lista.append("na")
        else:    
            lista.append(locations[i].findAll("td")[2].text)
    
    print(pd.DataFrame(lista))
```
